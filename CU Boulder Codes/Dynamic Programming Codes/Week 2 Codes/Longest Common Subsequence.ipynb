```python
def lcss(s1, s2, i, j):
    if i >= len(s1):
        return 0
    elif j >= len(s2):
        return 0
    else:
        if s1[i] == s2[j]:
            return 1 + lcss(s1,s2,i+1,j+1)
        else:
            opt1 = lcss(s1,s2,i+1, j)
            opt2 = lcss(s1,s2,i,j+1)
            return max(opt1,opt2)
```

```python
s1 = "GATTACA"
s2 = "ACTGATAACAA"
print(lcss(s1, s2, 0, 0))
```

```python
#slightly longer strings will take "forever" to run
s1 = "GGATTACACATTACCTATAGGTATAAT"
s2 = "GGATTTATCTATAAATTACCTATTTATTATATTACCGTATGGTATGC"
print(lcss(s1, s2, 0, 0))
```

```python
def lcssMemo(s1,s2):
    m = len(s1)
    n = len(s2)
    memoTable = [[0 for i in range(n+1)] for j in range(m+1)]
    decisionTable = [['' for i in range(n+1)] for j in range(m+1)]
    
    for i in range(m-1, -1, -1):
        for j in range(n-1, -1, -1):
            if s1[i] == s2[j]:
                memoTable[i][j] = 1 + memoTable[i+1][j+1]
                decisionTable[i][j] = "match"
            else:
                opt1 = memoTable[i+1][j]
                opt2 = memoTable[i][j+1]
                if opt1 > opt2:
                    memoTable[i][j] = memoTable[i+1][j]
                    decisionTable[i][j] = 'right'
                else:
                    memoTable[i][j] = memoTable[i][j+1]
                    decisionTable[i][j] = 'down'
    
    i = 0
    j = 0
    substr = ""
    while (i < m and j < n):
        if decisionTable[i][j] == "match":
            substr = substr + s1[i]
            i, j = i + 1, j + 1
        elif decisionTable[i][j] == 'right':
            i = i + 1
        elif decisionTable[i][j] == 'down':
            j = j + 1
    return memoTable[0][0], substr
```

```python
s1 = "GATTACA"
s2 = "ACTGATAACAA"
lcssMemo(s1,s2)
```
