```python
sizes = [1,3,5,10,20,50,75]
prices = [.1, .2, .4, .9, 3.1, 5.1, 8.2]
length = 100

def maxRevenue(L, sizes, prices):
    if (L == 0):
        return 0
    elif (len(sizes) != len(prices)):
        return None
    elif (L < 0):
        return (-10000000)
    else:
        maxRevList = [prices[i] + maxRevenue(L - sizes[i] + prices[i], sizes, prices) for i in range(len(sizes))]
        maxRevRolling = max(maxRevList)
        return maxRevRolling

maxRevenue(length, sizes, prices)
```

```python
def maxRevenue_Recursive(L, sizes, prices):
    if L == 0: 
        return 0
    if L < 0: 
        return (-100000000) # Just a large negative number will do
    k = len(sizes)
    assert len(prices) == k
    # Let us implement the max of 
    optionValues = [ (prices[i] + maxRevenue_Recursive(L-sizes[i], sizes, prices)) for i in range(k) ]
    optionValues.append(0) # also add 0 to cover the case where we waste 
    bestValueSoFar = max(optionValues)
    return bestValueSoFar

maxRevenue_Recursive(length, sizes, prices)
```

```python
def maxRevenue_Memoize(L, sizes, prices): 
    T = [0]*(L+1) # create an array of size L+1 and fill it with all 0s 
    k = len(sizes)
    assert len(prices) == k
    
    for l in range(1, L+1):
        optionValues = [ (prices[i] + T[l-sizes[i]]) for i in range(k) if l - sizes[i] >= 0 ]
        optionValues.append(0)
        T[l] = max(optionValues)
    return T[L]
```

```python
def maxRevenueMemo(L, sizes, prices):
    memoTable = [0]*(L+1)
    assert len(sizes) == len(prices)
    
    for l in range(1, L + 1):
        priceList = [(prices[i] + memoTable[l - sizes[i]]) for i in range(len(sizes)) if l - sizes[i] >= 0]
        priceList.append(0)
        memoTable[l] = max(priceList)
    return memoTable[L]
```

```python
sizes = [1,3,5,10,20,50,75]
prices = [.1, .2, .4, .9, 3.1, 5.1, 8.2]

print(maxRevenueMemo(30, sizes, prices))
```

```python
def maxRevenueMemoSoln(L, sizes, prices):
    memoTable = [0] * (L + 1)
    decisionMade = [0] * (L + 1) 
    
    assert len(sizes) == len(prices)
    
    for l in range(1, 1 + L):
        for j in range(len(sizes)):
            if l - sizes[j] >= 0:
                val = prices[j] + memoTable[l-sizes[j]]
                if val > memoTable[l]:
                    memoTable[l] = val
                    decisionMade[l] = j
                
    pathLength = L
    optimalPath = []
    currentVal = L
    
    while pathLength > 0:
        pathLength = pathLength - sizes[decisionMade[currentVal]]
        optimalPath.append(sizes[decisionMade[currentVal]])
        currentVal = pathLength
        
    return memoTable, optimalPath
```
