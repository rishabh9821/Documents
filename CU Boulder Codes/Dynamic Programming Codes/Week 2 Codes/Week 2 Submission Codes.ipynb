```python
# Program the recurrence for longest stable subsequence
# 0 <= i <= len(a)
# Note that if j == -1, then take aj = None
# else aj = a[j]
def lssLength(a, i, j):
    aj = a[j] if 0 <= j < len(a) else None 
    assert j < i
    if i == 0 and j == -1:
        lssLength(a,1,0)
    if i == len(a):
        return 1
    if abs(a[i] - a[j]) > 1:
        return lssLength(a,i+1,j)
    else:
        opt1 = 1 + lssLength(a,i+1,i)
        if j + 1 < i:
            opt2 = lssLength(a,i,j + 1)
        else:
            opt2 = 0
        opt3 = lssLength(a,i+1, j)
        return max(opt1,opt2, opt3)
```
