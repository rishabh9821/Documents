```python
df_povT = df_pov.T
df_povT.columns = df_povT.iloc[0,:]
df_povT = df_povT.iloc[1:,:]
df_povT
```

```python
df_povT.rename(index = {"Characteristic":"Year"}, inplace = True)
df_povT.columns = ['Total in the labor force','In poverty','Working poor rate for labor force',"Unrelated Individuals",'In poverty',"Working poor rate for Unrelated Families","Primary Families",'In poverty',"Working poor rate for primary families"]
df_povT
```

```python
plt.figure(figsize = (10,8))
df_povT = df_povT.astype('int')
sns.regplot(data = df_povT, x = df_povT.index, y = "Working poor rate for labor force")
plt.title('Year vs. Working Poor Rate for Labor Force')
plt.savefig('Poor Rate.jpeg')
```

```python
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(df_povT.index,df_povT["Working poor rate for labor force"])
```

```python
print('The formula of the graph follows this formula')
print(slope,"(year)","+",intercept)
print("-------------------------------------------------------------------")
print("This formula takes an r2 - value of approximately",r_value**2)
print("-------------------------------------------------------------------")
print("With this model, we can predict a ",round(slope*2024 + intercept,2),"% poverty rate among working class citizens in 2024 barring any strong changes")
```

```python
import os

directory = "/Users/rishabhagarwal/Downloads/Industry over Last 10 Years"
pathList = list()
for path in os.listdir(directory):
    filePath = os.path.join(directory, path)
    pathList.append(filePath)
    
pathList
```

```python
import pandas as pd

i = 0
for path in pathList:
    i = i + 1
    df = pd.read_excel(path)
    name = df.iloc[8,1]
    df = df.iloc[18:,:]
    strng = "Avg. # of Workers for "+name
    df.columns = ['Year',"Qtr1","Qtr2","Qtr3","Qtr4"]
    df = df.iloc[1:,:]
    df[strng] = (df['Qtr1'] + df["Qtr2"] + df['Qtr3'] + df["Qtr4"])/4
    df.dropna(inplace = True)
    df.set_index("Year", inplace = True)
    df = pd.DataFrame(df.iloc[:,4])
    if i == 1:
        tot_df = df
    else:
        tot_df = pd.concat([tot_df,df],axis = 1)
        
tot_df
```
