```python
import pandas as pd

df = pd.read_excel('~/Downloads/cpsaat39.xlsx')
df = df[4:]
df.set_index('')
```

```python
lst = df.loc[4,:].values
lst = lst[~pd.isnull(lst)]
lst = list(lst)
lst
```

```python
tot_headers = ["Occupation"] + lst
df.columns = tot_headers
df.head(10)
```

```python
df = df.loc[6:,:]
df.drop([7], inplace = True)
df
```

```python
df.dropna(subset = ["Occupation"], inplace = True)
df.set_index('Occupation', inplace = True)
df.head()
```

```python
cols = df.columns
new_col_list = list()
i = 0
for col in cols:
    if i in (0,1):
        new_col_list.append(col+" (Total)")
    elif i in (2,3):
        new_col_list.append(col+" (Male)")
    else:
        new_col_list.append(col+" (Female)")
    i = i+1
df.columns = new_col_list
df.head()
```

```python
df['Number of Worker Discrepancy'] = df['Number of workers (Male)'] - df['Number of workers (Female)']
df.head()
df2 = df.sort_values(by = "Number of Worker Discrepancy", ascending = False)
df2.dropna(inplace = True)
df2
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df3 = df2.iloc[0:5,:]

plt.figure(figsize = (12,10))
sns.barplot(x = df3.index, y = "Number of Worker Discrepancy", data = df3)
ax = sns.countplot(x=df3.index, data=df3)
plt.ylabel("Number of Worker Difference Between Male & Female")
plt.title('Jobs where the Difference between Male & Female Workers are Highest')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
df.replace('–', np.nan, inplace = True)
df.dropna(inplace = True)
df = df.astype('int')
df['Avg. Difference in Male & Female Salaries'] = (df['Median weekly earnings (Male)'] - df['Median weekly earnings (Female)'])*52
df.head()
```

```python
df2 = df[df['Median weekly earnings (Male)'] > 1000]
df2 = df2[df2['Median weekly earnings (Female)'] > 1000]
df2.sort_values(by = "Avg. Difference in Male & Female Salaries", ascending = False)
```

```python
df3 = df2.iloc[0:5,:]

plt.figure(figsize = (12,10))
sns.barplot(x = df3.index, y = "Avg. Difference in Male & Female Salaries", data = df3)
# ax = sns.countplot(x=df3.index, data=df3)
plt.ylabel("Avg. Difference in Male & Female Salaries")
plt.title('Jobs where the Salary Difference between Male & Female are the Highest')
# ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
# plt.tight_layout()
# plt.show()
```

```python
df_pov = pd.read_excel("~/Downloads/working-poor-tables-2019.xlsx")
df_pov = df_pov.iloc[2:11,:13]
df_pov.columns = ['Characteristic'] + list(range(2008,2020))
df_pov
```

```python
df_povT = df_pov.T
df_povT.columns = df_povT.iloc[0,:]
df_povT = df_povT.iloc[1:,:]
df_povT
```

```python
df_povT.rename(index = {"Characteristic":"Year"}, inplace = True)
df_povT.columns = ['Total in the labor force','In poverty','Working poor rate for labor force',"Unrelated Individuals",'In poverty',"Working poor rate for Unrelated Families","Primary Families",'In poverty',"Working poor rate for primary families"]
df_povT
```

```python
plt.figure(figsize = (10,8))
df_povT = df_povT.astype('int')
sns.regplot(data = df_povT, x = df_povT.index, y = "Working poor rate for labor force")
plt.title('Year vs. Working Poor Rate for Labor Force')
```

```python
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(df_povT.index,df_povT["Working poor rate for labor force"])
```

```python
print('The formula of the graph follows this formula')
print(slope,"(year)","+",intercept)
print("-------------------------------------------------------------------")
print("This formula takes an r2 - value of approximately",r_value**2)
print("-------------------------------------------------------------------")
print("With this model, we can predict a ",slope*2024 + intercept,"% poverty rate among working class citizens barring any strong changes")
```
